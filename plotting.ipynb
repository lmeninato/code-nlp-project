```python
import pandas as pd
import matplotlib.pyplot as plt

def extract_model_name(filename):
    return filename.split("/")[1]

def plot_csv_files(filenames, title_of_metric):
    for filename in filenames:
        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(filename)

        # Plot the 'Step' column on the x-axis and the 'Value' column on the y-axis
        plt.plot(df["Step"], df["Value"], label=extract_model_name(filename))

    # Set the labels and the title for the plot
    plt.xlabel("Training Step")
    plt.title(title_of_metric)

    # Display the legend
    plt.legend()

    # Show the plot
    plt.show()


```

```python
# Example usage
bert_small_model_path = 'results/bert-small-py3/run-logs-tag-'
flan_t5_small_model_path = 'results/flan-t5-small-py3/run-logs-tag-'
t5_small_model_path = 'results/t5-small-py3/run-logs-tag-'

metrics = [("Evaluation BLEU Score", 'eval_bleu'), ("Training Loss", 'train_loss'),
          ("Evaluation ROUGE-1 Score", 'eval_rouge1'), ("Evaluation ROUGE-2 Score", 'eval_rouge2'),
          ("Evaluation Loss", 'eval_loss')]

for title, tag in metrics:
    csv_files = [f'{bert_small_model_path}{tag}.csv', f'{flan_t5_small_model_path}{tag}.csv', f'{t5_small_model_path}{tag}.csv']
    plot_csv_files(csv_files, title)
```
